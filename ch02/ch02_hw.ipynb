<p>(e) Use the boxplot() method of college to produce side-by-side
boxplots of Outstate versus Private.</p>

```python
college['Private'] = college['Private'].replace('Yes', 'Private Colleges')
college['Private'] = college['Private'].replace('No', 'Public Colleges')
```

<p>(b) & (c) What is the range, mean and standard deviation of each quantitative predictor? </p>


```python
for pred in quantitative_predictors:
    print(f'{pred} min: {auto_stats[pred].min()}')
    print(f'{pred} max: {auto_stats[pred].max()}')
    print(f'{pred} mean: {auto_stats[pred].mean()}')
    print(f'{pred} standard deviation: {auto_stats[pred].std()}')
    print()
```

```python
remaining_auto_data = auto.drop(drop_rows)
remaining_auto_data.head(15)
```

```python
remaining_auto_data_stats = remaining_auto_data.describe()
remaining_auto_data_stats
```

```python
for pred in quantitative_predictors:
    print(f'{pred} min: {remaining_auto_data_stats[pred].min()}')
    print(f'{pred} max: {remaining_auto_data_stats[pred].max()}')
    print(f'{pred} mean: {remaining_auto_data_stats[pred].mean()}')
    print(f'{pred} standard deviation: {remaining_auto_data_stats[pred].std()}')
    print()
```
