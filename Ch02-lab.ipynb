## Basics of Selecting Rows and Columns

```python
Auto.columns
```

```python
Auto[:3]
```

```python
Auto[Auto['year'] > 80]
```

```python
 Auto[['mpg', 'horsepower']]
```

```python
Auto.index
```

```python
Auto_re = Auto.set_index('name')
Auto_re
```

```python
Auto_re.columns
```

```python
# loc selects data using row and column names
# iloc uses numerical indices
rows = ['amc rebel sst', 'ford torino']
Auto_re.loc[rows]
```

```python
# retrieve the 4th and 5th rows
Auto_re.iloc[[3,4]]
```

```python
# retrieve the 1st, 3rd and and 4th columns
Auto_re.iloc[[0,2,3]]
```

```python
# extract the 4th and 5th rows, as well as the 1st, 3rd and 4th columns
Auto_re.iloc[[3,4],[0,2,3]]
```

```python
Auto_re.loc['ford galaxie 500', ['mpg', 'origin']]
```

## More on Selecting Rows and Columns

```python
idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80, ['weight', 'origin']]
```

```python
 Auto_re.loc[lambda df: df['year'] == 80, ['weight', 'origin']]
```
