```python
a_inverse = np.linalg.inv(a)
b_inverse = np.linalg.inv(b)
a_and_b_inverse = np.linalg.inv(np.matmul(a, b))
```

```python
a_and_b_inverse
```

```python
a_and_b_inverse = np.linalg.inv(a_and_b)
a_and_b_inverse
```

### 2.4.1 Groups

1. Closure of G under ⊗: ∀x, y ∈ G : x ⊗ y ∈ G \
&emsp;&emsp;For all x, y under the set G, the output of x+y or x*y is the under the set G 
2. Associativity: ∀x, y, z ∈ G : (x ⊗ y) ⊗ z = x ⊗ (y ⊗ z) \
&emsp;&emsp;for x, y, z under the set G, (x+y)+(z)==x+(y+x), (xy)z==x(yz)
3. Neutral element: ∃e ∈ G ∀x ∈ G : x ⊗ e = x and e ⊗ x = x \
&emsp;&emsp;There exist a 'e" where x+e=e and x*e=x
4. Inverse element: ∀x ∈ G ∃y ∈ G : x ⊗ y = e and y ⊗ x = e, where e is the neutral element. We often write $x^{−1}$ to denote the inverse element of x. \
&emsp;&emsp;For x under the set G, there exist y where xy = e

### 2.1 consider (R\{−1}, ⋆), where a ⋆ b := ab + a + b, a, b ∈ R\{−1} 
**a. Show that (R\{−1}, ⋆) is an Abelian group** \
&emsp;&emsp;Note: operation a*b should never produce -1 \
&emsp;&emsp;Since a\*b$\neq$-1, we can rewrite the operation as a ⋆ b := ab + a + b + 1 where a$\neq$-1 and b$\neq$-1

**Abelian group criteria** \
1. Closure:\
&emsp;&emsp;a, b ∈ R \
&emsp;&emsp;ab + a + b + 1  ∈ R \
&emsp;&emsp;Closure condition met 
2. Neutral element: \
&emsp;&emsp;Neutral element=0
4. Associatiity: \
&emsp;&emsp;a, b, c ∈ R \
&emsp;&emsp;(ab + a + b) + c = ab + (a + b + c) \
&emsp;&emsp;Associativity condition met 
5. Neutral/Identity element: \
&emsp;&emsp; ab + a + b + 0 = ab + a + b \
&emsp;&emsp; Neutral element condition met  
6. Inverse element: inverse condition met 
7. a⋆b=ab+a+b=ba+b+a=b⋆a \
&emsp;&emsp;Commutative condition met \
thus it's an Abelian group
