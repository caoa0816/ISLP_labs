```python
import numpy as np
import scipy as sp
import scipy.special as sps
import matplotlib.pyplot as plt
```

```python
n = 100
p = 0.5
q = 1-p
```

## Binomial Distribution 

```python
binomial_distribution = []
```

```python
for k in range(n+1):
    probability = sps.comb(n, k) * p**k * q**(n-k)
    binomial_distribution.append(probability)
```

## Normal distribution (Approximation)

**mu: mean** \
**sigma: variance, aka standard deviation squared**

```python
mu = n*p
sigma = np.sqrt(n*p*q)
```

```python
x = np.arange(0, n, 1)
```

```python
coefficient = 1 / (sigma * np.sqrt(2 * np.pi))
exponent = -((x - mu) ** 2) / (2.0 * sigma ** 2)
normal_distribution = coefficient * np.exp(exponent)
```

## Plot how binomial & normal distribution converges 

```python
plt.plot(binomial_distribution, label="binomial")
plt.plot(normal_distribution, label="normal")
plt.legend()
plt.xlim(0, n)
```
