<h1>Vector arithmatic</h1>
<p>Addition, subtraction, multiplication & division can only be performed on vectors of the same dimensions</p>

```python
a = np.array([24, 50, 89])
b = np.array([21, 10, 8])
c = np.array([34, 70, 9, 67])
```

```python
a+b
```

```python
a-b
```

```python
a*b
```

```python
a/b
```
