<h1>2.Dot product of vectors</h1>

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])
```

```python
np.dot(a, b)
```

<h3>Dot product is commutative</h3>
<pre>
    a * b = b * a
</pre>
<p>example below</p>

```python
#np.dot(a, b) == np.dot(b, a)
np.dot(b, a)
```

<h3>Dot product is distributive</h3>
<pre>
    a(b+c) = a*b + a*c
</pre>

```python
c = np.array([11, 12, 13, 14, 15])
```

```python
np.dot(a, b+c)
```

```python
np.dot(a,b)+np.dot(a,c)
```
