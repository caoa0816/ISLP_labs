# 6. Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a nonparametric approach)? What are its disadvantages?

### Answer
Parametric approaches reduce the estimation problem to the estimation of few parameters by making assumptions about the parameters that will impact the final estimaton. <br>
Parametric approaches advantage: easy implemention & only needs small number of dataset<br>
Parametric approaches disadvantage: estimation might not reflect the true form of function<br>
<br>
Non-parametric approach doesn't make any assumptions about the form of the function f, so the estimation will be more accurate<br>
Non-parametric approach disadvantage: more data needed & overfitting of training data
